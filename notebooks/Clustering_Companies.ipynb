```python
#sentence = "k-means clustering has been used as a feature learning (or dictionary learning) step, in either (semi-)\
#supervised learning or unsupervised learning.[34] The basic approach is first to train a k-means clustering representation, \
#using the input training data (which need not be labelled). \
#Then, to project any input datum into the new feature space, we have a choice of encoding functions, but we can use for \
#example the thresholded matrix-product of the datum with the centroid locations, the distance from the datum to each centroid, \
#or simplyHi an indicator function for the nearest centroid,[34][35] or some smooth transformation of the distance.[36]\
#Alternatively, by transforming the sample-cluster distance through a Gaussian RBF, one effectively obtains the hidden \
#layer of a radial basis function network"


words = industry_text.split("~")
words = np.asarray(words) #So that indexing with a list will work
lev_similarity = -1*np.array([[distance.levenshtein(w1,w2) for w1 in words] for w2 in words])

affprop = sklearn.cluster.AffinityPropagation(affinity="precomputed", damping=0.7)
affprop.fit(lev_similarity)
cluster_num = 0
for cluster_id in np.unique(affprop.labels_):
    exemplar = words[affprop.cluster_centers_indices_[cluster_id]]
    cluster = np.unique(words[np.nonzero(affprop.labels_==cluster_id)])
    cluster_str = ", ".join(cluster)
    print("cluster - %d *%s:* %s" % (cluster_num, exemplar, cluster_str))
    cluster_num = cluster_num+1
```
